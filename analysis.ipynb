```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('charcters_stats.csv')
data
```

```python
data.shape
```

```python
data['Alignment'].value_counts()
```

```python
good_characters = data[data['Alignment']=='good']
good_characters.shape
```

```python
good_characters.sort_values(by='Intelligence', ascending=False)
```

```python
fastest_good_characters = good_characters[good_characters['Speed']==100]
fastest_good_characters.sort_values(by='Total', ascending=0)
```

```python
most_powerfull_good_characters = good_characters[good_characters['Power']==100]
print(most_powerfull_good_characters.shape[0])
most_powerfull_good_characters.sort_values(by='Total', ascending=0)
```

```python
best_good_characters = good_characters.sort_values(by='Total', ascending=False)
best_good_characters
```

```python
plt.figure(figsize=(16,10))
plt.bar(list(fastest_good_characters['Name']), list(fastest_good_characters['Total']), color='blue')
plt.rcParams.update({'font.size': 6})
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(most_powerfull_good_characters['Name'][0:5]), list(most_powerfull_good_characters['Total'][0:5]), color=[ 'orange', 'red','pink', 'purple', 'blue'])
plt.rcParams.update({'font.size': 10})
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(good_characters['Speed'])
plt.title('Distribution of Speed')
plt.xlabel('Speed')
plt.rcParams.update({'font.size': 10})
plt.show()
```
